```python
# Select the top 10 countries with the most gold medals
top10_countries = data.groupby("Country", sort=False)["Medal"].apply( lambda x: x.str.contains("Gold").sum()).nlargest(10)
# Create a bar chart
fig_top10, ax_top10 = plt.subplots(figsize=(10, 6))

# Customize the plot
ax_top10.bar(top10_countries.index, top10_countries.values, color='#005f73')
ax_top10.set_facecolor('#EAF4f4')
ax_top10.spines['top'].set_visible(False)
ax_top10.spines['right'].set_visible(False)
ax_top10.spines['left'].set_visible(False)
ax_top10.set_ylabel('Number of Gold Medals', fontsize=10)
ax_top10.set_title('Top 10 Countries with the Most Gold Medals', fontsize=12, color='#22333b')
ax_top10.tick_params(length=0, labelrotation=45, colors='#22333b')

# Show the plot
plt.show()
```

```python
# Group the data by year and gender and calculate the sum of medals
gender_year_data = data.groupby(['Year', 'Gender'])['Medal'].count().unstack()

# Create a line chart
fig_gen, ax_gen = plt.subplots(figsize=(10, 6))

# Customize the plot
ax_gen.plot(gender_year_data.index, gender_year_data['Men'], color='#005f73', label='Men', marker = 'o')
ax_gen.plot(gender_year_data.index, gender_year_data['Women'], color='#94d2bd', label='Women', marker = 'o')
ax_gen.fill_between(gender_year_data.index, gender_year_data['Men'],alpha = 0.6 , color='#005f73')
ax_gen.fill_between(gender_year_data.index, gender_year_data['Women'],alpha = 0.6 , color='#94d2bd')
ax_gen.set_facecolor('#EAF4f4')
ax_gen.spines['top'].set_visible(False)
ax_gen.spines['right'].set_visible(False)
ax_gen.spines['left'].set_visible(False)
ax_gen.set_xlim([1896,2012])
ax_gen.yaxis.tick_right()
ax_gen.set_ylabel('Number of Won Medals', fontsize=10, color='#22333b')
ax_gen.set_title('Male and Female Won Medals over the Years', fontsize=12, color='#22333b')
ax_gen.tick_params(length=0, colors='#22333b')
ax_gen.legend()

# Show the plot
plt.show()
```

```python
# Select the data for Slovakia
slovakia_data = data[data['Country'] == 'SVK']

# Group the data by sport and calculate the count of medals
medals_by_sport = slovakia_data.groupby('Sport')['Medal'].count()

# Create the outer donut chart
fig_SVK, ax_SVK = plt.subplots(figsize=(10, 8))
wedge_colors = ['#d8f3dc', '#b7e4c7', '#95d5b2', '#74c69d', '#52b788', '#40916c']
ax_SVK.pie(medals_by_sport.values, labels=medals_by_sport.index, autopct='%1.1f%%', colors=wedge_colors, radius=1, wedgeprops=dict(width=0.3, edgecolor='w'))

# Create the inner donut chart
medals_by_type = slovakia_data.groupby('Sport')['Medal'].value_counts()
inner_colors = {'Gold': '#E9C46A', 'Silver': '#E0E1DD', 'Bronze': '#CD7F32'}
ax_SVK.pie(medals_by_type, radius=0.7, autopct='%1.1f%%', pctdistance=0.75, colors=[inner_colors.get(x[1]) for x in medals_by_type.index])

# Customize the plot
ax_SVK.set_title('Count of Medals in each Sport', fontsize=12, color='#22333b')

# Show the plot
plt.show()
```
